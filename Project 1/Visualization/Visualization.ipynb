## Subject visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from pathlib import Path as p

data_path = "D:\BitBucket/5280\Project\Data\Engineered"
df = pd.read_parquet(p(data_path) / 'combined_df_no_hw.parquet')
```

## Subject visualization and analysis

```python
subject_info = df.drop_duplicates(subset=['age', 'gender', 'bmi'])[['age', 'gender', 'bmi']]
subject_info.head()
```

```python
plt.figure(figsize=(10, 12))
# Box plot of age
plt.subplot(311)
sb.boxplot(x=subject_info['age'])

# Count of gender
plt.subplot(312)
sb.countplot(x=subject_info['gender'])
# Box plot of bmi
plt.subplot(313)
sb.boxplot(x=subject_info['bmi'])
```

It can be seen that participants are at their prime time with the age average falls about 28, the youngest is 18 and the eldest is 35, which indicates that they fit criteria of the event.
Gender is not so imbalanced (10 /14), which is acceptable
Even though bmi is a controversial metric of telling how healthy humans are, it's a useful reference tool. Again, 50% of subjects have good and satisfactory bmi

```python
sb.heatmap(subject_info.corr(), annot=True, cmap='Blues')
```

It's interesting that age and gender's correlation is even stronger that age vs bmi since human bodies tend to become unhealthier when we become elder. We deem that the result comes from a tiny dataset of 24.
